```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data
```

## Player Count

* Display the total number of players


```python
#find total number of players using unique so no duplicate SN
unique_players=purchase_data["SN"].unique()
total_number_players=(len(unique_players))
print(total_number_players)
```

## Purchasing Analysis (Total)

```python
#drop duplicated by using unique() and use len() to get total number of unique items
unique_items=purchase_data["Item ID"].unique()
num_un_items=(len(unique_items))

#create df with duplicate IDs dropped using .drop_duplicate
dup_id_dropped_df=purchase_data.drop_duplicates(subset="Item ID", keep="last")

#calculate average price, total revenue
average_price=(dup_id_dropped_df["Price"]).mean()
total_rev=purchase_data["Price"].sum()
dup_id_dropped_df["Price"].describe()

#create a dict for summary table, make dict into df and print with rounding to 2 decimal places
summary_dict=[{"Number of Unique Items": num_un_items, "Average Price": average_price,"Number of Purchase":len(purchase_data), "Total Revenue":total_rev}]
summary_df=(pd.DataFrame(summary_dict)).round(2)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#drop duplicates in SN
no_dupl_df=purchase_data.drop_duplicates(subset=["SN"], keep='first')

#count females from no_dupl and add to list 
female_count=(no_dupl_df['Gender'].tolist().count('Female'))

#count males from no_dupl and add to list
male_count=(no_dupl_df['Gender'].tolist().count('Male'))

#add together male and female count and subtract from total players from unique to get other count
other_count=total_number_players-(male_count+female_count)

#create gender dict
gender_count={"":["Male","Female","Other/Non-Disclosed"], 
        "Total Count":[male_count,female_count,other_count], 
        "Percentage of Players":[(100*(male_count/total_number_players)),(100*(female_count/total_number_players)),(100*(other_count/total_number_players))]}

#create gender df 
gender_count_df=pd.DataFrame(gender_count,columns= ["", "Total Count", "Percentage of Players"])

#format percentage column
gender_count_df['Percentage of Players'] = gender_count_df['Percentage of Players'].apply('{:.2f}%'.format)

#clean up
gender_summary=gender_count_df.rename(columns = {"":"Gender"})

gender_summary.set_index(["Gender"])


```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# purchase count by gender
fpurchase_count=(purchase_data['Gender'].tolist().count('Female'))
mpurchase_count=(purchase_data['Gender'].tolist().count('Male'))
opurchase_count=len(purchase_data)-(fpurchase_count+mpurchase_count)

#avg purchase price: use groupby gender to find mean of "price" 
avg_purchase_price=(purchase_data.groupby(by='Gender')['Price'].mean()).round(2)

#avg purchase total per person per gender
total_purchase=pd.DataFrame(purchase_data.groupby(by="Gender")['Price'].sum())
favg_purchase =(((total_purchase.loc["Female","Price"])/female_count))
mavg_purchase=(((total_purchase.loc["Male","Price"])/male_count))           
oavg_purchase=((total_purchase.loc["Other / Non-Disclosed","Price"])/other_count)

#summary table columns: purchase count, Average purchase price, average purchase total
summary_table = {"Gender":["Male","Female","Other/Non-Disclosed"],
               "Purchase Count":[mpurchase_count, fpurchase_count, opurchase_count], 
                "Average Purchase Price":[avg_purchase_price.loc["Male"], avg_purchase_price.loc["Female"],avg_purchase_price.loc["Other / Non-Disclosed"]],
                "Average Purchase Total": [mavg_purchase,favg_purchase, oavg_purchase],
                "Total Purchase Value": [total_purchase.loc["Male","Price"], total_purchase.loc["Female","Price"], total_purchase.loc["Other / Non-Disclosed","Price"]]}
summary=pd.DataFrame(summary_table)
summary.round(2)
summary_index=summary.set_index(["Gender"])
summary_index
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#drop duplicate SN 
no_dupl_sn=pd.DataFrame(purchase_data.drop_duplicates(subset="SN", keep="first"))
#create bins for ages min=7, max =45, avg=30
bins =[7,9.9,14.9,19.9,24.9,29.9,34.9,39.9,50]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
no_dupl_sn["Age Ranges"]=pd.cut(no_dupl_sn["Age"], bins, labels=group_names, include_lowest=True)

age_count= no_dupl_sn["Age Ranges"].value_counts()

#sort by age group
age_count_unsorted=pd.DataFrame(age_count)
age_count_sorted=pd.DataFrame(age_count_unsorted.sort_index())

#age purchase percentages
age_perc=pd.DataFrame((age_count/(len(no_dupl_sn)))*100)
age_count_per=pd.DataFrame(age_perc.sort_index())
#format
age_count_format=pd.DataFrame(age_count_per["Age Ranges"].map("{:.2f}%".format))
#Create Summary Table: Total Count and Percentage of players by age using age_count and age_count_per

age_count_sorted['Percentage of Players']= age_count_format['Age Ranges']
age_count=age_count_sorted
age_count.rename(columns={"Age Ranges": "Total Count"})
```
